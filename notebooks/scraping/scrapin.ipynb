```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
from pathlib import Path
import re
```

```python
def get_pdfs_from_page(url, download=False, output_dir="pdfs"):
    """
    Extract all PDF links from a webpage.

    Parameters
    ----------
    url : str
        URL of the webpage.
    download : bool, default=False
        Whether to download the PDFs.
    output_dir : str, default="pdfs"
        Directory where PDFs will be saved.

    Returns
    -------
    pd.DataFrame
        DataFrame with columns: name, url
    """

    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    pdfs = []

    for a in soup.find_all("a", href=True):
        full_url = urljoin(url, a["href"])

        if ".pdf" in full_url.lower():
            name = a.get_text(strip=True)

            # Fallback to filename if anchor text is empty
            if not name:
                name = full_url.split("/")[-1]

            pdfs.append({
                "name": name,
                "url": full_url
            })

    df = pd.DataFrame(pdfs).drop_duplicates()

    if download and not df.empty:
        out_dir = Path(output_dir)
        out_dir.mkdir(parents=True, exist_ok=True)

        for _, row in df.iterrows():
            # Remove invalid filename characters
            safe_name = re.sub(r'[<>:"/\\|?*]', "_", row["name"])
            filename = out_dir / f"{safe_name}.pdf"

            r = requests.get(row["url"], headers=headers)
            r.raise_for_status()

            with open(filename, "wb") as f:
                f.write(r.content)

            print(f"Downloaded: {filename}")

    return df
```
