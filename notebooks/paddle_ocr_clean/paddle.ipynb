```python
class JSONWriter:
    @staticmethod
    def write(text: str, path: str) -> None:
        """
        Writer for default JSON input to label studio

        Only text data
        """
        text = html.escape(text)

        task_dict = {
            "data": {
                "text": text
            }
        }
        task = [task_dict]
        
        with open(path, "w", encoding="utf-8") as f:
            json.dump(task, f, ensure_ascii=False, indent=2)
```

```python
months = {
    "stycznia": 1,
    "lutego": 2,
    "marca": 3,
    "kwietnia": 4,
    "maja": 5,
    "czerwca": 6,
    "lipca": 7,
    "sierpnia": 8,
    "wrzesnia": 9,
    "września": 9,
    "pazdziernika": 10,
    "października": 10,
    "listopada": 11,
    "grudnia": 12,
}
```

```python
def get_metadata(reg_name):
    m = re.search(
        r"nr[-_ ]*(\d+)",
        reg_name,
        re.IGNORECASE
    )

    date_match = re.search(
        r"z-dnia-(\d{1,2})-([a-ząćęłńóśźż]+)-(\d{4})",
        reg_name,
        re.IGNORECASE
    )

    metadata = {}

    if m:
        metadata["regulation_no"] = int(m.group(1))

    if date_match:
        day = int(date_match.group(1))
        month_name = date_match.group(2).lower()
        year = int(date_match.group(3))

        month_num = months.get(month_name)

        if month_num:
            metadata["date"] = (
                f"{year}-{month_num:02d}-{day:02d}"
            )
    return metadata
```

```python
def run_ocr(ocr: PaddleOCR, path: str):

    result = ocr.predict(path)
    txts = []
    for res in result:
        txts.append("\n".join(res["rec_texts"]))
    
    whole_text = "\n\n##### nowa_strona #####\n\n".join(txts)
    whole_text = html.escape(whole_text)

    return whole_text
```

```python
root = Path("./wazne_pdfy")
output_file = root / "label_studio_tasks.json"
with open(output_file, "w", encoding="utf-8") as out:
    out.write("[")
    first = True
    for pdf_file in root.glob("*.pdf"):
        
        whole_text = run_ocr(ocr=ocr, path=str(pdf_file.resolve()))
        task = {
                            "data": {
                                "text": whole_text,
                                "pdf_path": str(pdf_file.name),
                                "year": -1,
                                "number": -1,
                                "date": -1
                            }
                        }
        if not first:
            out.write(",\n")

        json.dump(task, out, ensure_ascii=False)
        first = False
        print(pdf_file)
    out.write("]")
```

```python
root = Path("../Parsowanie_pdf/pdfs_one_source")

for year_dir in sorted(root.iterdir()):

    if not year_dir.is_dir():
        continue

    output_file = year_dir / "label_studio_tasks.json"

    with open(output_file, "w", encoding="utf-8") as out:

        out.write("[")
        first = True

        for regulation_dir in sorted(year_dir.iterdir()):
            if not regulation_dir.is_dir():
                continue

            for pdf_file in regulation_dir.glob("*.pdf"):
                whole_text = run_ocr(ocr=ocr, path=str(pdf_file.resolve()))

                reg_name = pdf_file.parent.name
                metadata = get_metadata(reg_name)

                print(metadata)

                task = {
                            "data": {
                                "text": whole_text,
                                "pdf_path": str(pdf_file.name),
                                "year": year_dir.name,
                                "number": metadata.get("regulation_no", -1),
                                "date": metadata.get("date", -1)
                            }
                        }

                if not first:
                    out.write(",\n")

                json.dump(task, out, ensure_ascii=False)
                first = False

        out.write("]")
```

```python
import json
from pathlib import Path

root_dir = Path("../Parsowanie_pdf/pdfs_one_source")

for year_dir in root_dir.iterdir():
    if not year_dir.is_dir():
        continue

    json_file = year_dir / "label_studio_tasks.json"
    if not json_file.exists():
        continue

    print(f"Processing {year_dir.name}...")

    with open(json_file, "r", encoding="utf-8") as f:
        tasks = json.load(f)

    # Map PDF filename -> relative path within the year directory
    pdf_map = {
        pdf.name: str(pdf.relative_to(year_dir))
        for pdf in year_dir.rglob("*.pdf")
    }

    updated = 0
    for task in tasks:
        filename = task.get("data").get("pdf_path")
        print(filename)
        if filename in pdf_map:
            task["data"]["pdf_path"] = pdf_map[filename]
            updated += 1

    with open(json_file, "w", encoding="utf-8") as f:
        json.dump(tasks, f, ensure_ascii=False, indent=2)

    print(f"  Updated {updated} tasks")
```

```python
from pathlib import Path
import shutil

src_root = Path("../Parsowanie_pdf/pdfs_one_source")
dst_root = Path("zarzadzenia_text_json")

for year_dir in src_root.iterdir():
    if not year_dir.is_dir() or not year_dir.name.isdigit():
        continue

    src_json = year_dir / "label_studio_tasks.json"
    if not src_json.exists():
        continue

    dst_year = dst_root / year_dir.name
    dst_year.mkdir(parents=True, exist_ok=True)

    shutil.copy2(src_json, dst_year / "label_studio_tasks.json")

    print(f"Copied {src_json}")

print("Done.")
```
